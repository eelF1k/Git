1.Функції **all()** та **any()** - це вбудовані функції Python, які допомагають в роботі з ітерованими об'єктами, такими як списки, кортежі або інші послідовності. Вот як вони працюють:

1)**all(iterable)**:

*   Функція **all()** приймає один аргумент - ітерабельний об'єкт, такий як список або кортеж.
*   Вона повертає **True**, якщо всі елементи в ітерабельному об'єкті є істинними (не рівні **False**, **None**, **0**, пустий рядок тощо).
*   Якщо хоча б один елемент є ложним, функція **all()** повертає **False**.
Приклад використання **all()**:

```python
my_list = [True, True, True, False]
result = all(my_list)
print(result)  # Поверне False, оскільки є хоча б один False
```

2)**any(iterable)**:

*   Функція **any()** також приймає один аргумент - ітерабельний об'єкт.
*   Вона повертає **True**, якщо хоча б один елемент в ітерабельному об'єкті є істинним.
*   Якщо всі елементи є ложними, функція **any()** повертає **False**.

Приклад використання **any()**:

```python
my_list = [False, False, False, True]
result = any(my_list)
print(result)  # Поверне True, оскільки є хоча б один True
```

Створення таблиці в Python може бути виконано за допомогою різних бібліотек, але однією з найпоширеніших є бібліотека Pandas. Pandas надає зручні структури даних для роботи з таблицями, зокрема DataFrame. Ось як створити таблицю за допомогою Pandas:

```python
%%html
<!DOCTYPE html>
<html>
<head>
<style>
table {
  font-family: arial, sans-serif;
  border-collapse: collapse;
  width: 100%;
}

td, th {
  border: 1px solid #dddddd;
  text-align: left;
  padding: 8px;
}

tr:nth-child(even) {
  background-color: #dddddd;
}
</style>
</head>
<body>

<h2>HTML Table</h2>

<table>
  <tr>
    <th>Company</th>
    <th>Contact</th>
    <th>Country</th>
  </tr>
  <tr>
    <td>Alfreds Futterkiste</td>
    <td>Maria Anders</td>
    <td>Germany</td>
  </tr>
  <tr>
    <td>Centro comercial Moctezuma</td>
    <td>Francisco Chang</td>
    <td>Mexico</td>
  </tr>
  <tr>
    <td>Ernst Handel</td>
    <td>Roland Mendel</td>
    <td>Austria</td>
  </tr>
  <tr>
    <td>Island Trading</td>
    <td>Helen Bennett</td>
    <td>UK</td>
  </tr>
  <tr>
    <td>Laughing Bacchus Winecellars</td>
    <td>Yoshi Tannamuri</td>
    <td>Canada</td>
  </tr>
  <tr>
    <td>Magazzini Alimentari Riuniti</td>
    <td>Giovanni Rovelli</td>
    <td>Italy</td>
  </tr>
</table>

</body>
</html>

```

Цей код містить HTML-розмітку для відображення емодзі (піктограми) на веб-сторінці. Емодзі  і  представляють різні емодзі символи.

Ось пояснення коду:

*   **< p style="font-size:100px">&#128052 ;< /p>**:
*   **<p>** - це тег для створення абзацу (параграфу) в HTML.
*   **style="font-size:100px"** - ця атрибут style задає розмір шрифту для тексту в абзаці. У цьому випадку, розмір шрифту встановлено на 100 пікселів.
*   **&#128052**; - це HTML-код символу, який представляє конкретний емодзі. У цьому випадку, цей код відповідає емодзі "🐼".

**<p>I will display &#128052 ;</p>**:

*   Цей абзац просто відображає текст "I will display 🐼". Тут символ емодзі також вказаний за допомогою HTML-коду **&#128052**;.

**<p>I will display &#x1F434 ;</p>**:

*   Цей абзац також відображає текст "I will display 🐴". Тут символ емодзі вказаний за допомогою шістнадцяткового представлення **&#x1F434**;. У цьому випадку, цей код відповідає емодзі "🐴".


Отже, цей код використовує HTML для відображення емодзі на веб-сторінці, і ви можете контролювати їх розмір та відображення за допомогою відповідних атрибутів та кодів символів.

```python
%%html
<p style="font-size:100px">&#128052;</p>
<p>I will display &#128052;</p>
<p>I will display &#x1F434;</p>
```
